```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Loading data

# Lignes de X corresond aux images,
# colonnes de X correspond aux pixels.
X = np.load('../data/X.npy')
# y correspond aux classes.
y = np.load('../data/Y.npy')
img_size = 64

X = X.reshape(len(X), 64*64) # mise a plat de l image
```

```python
# Ã‰quivalence encodage/traduction
dic_equiv = {0:9, 1:0, 2:7, 3:6, 4:1, 5:8, 6:4, 7:3, 8:2 , 9:5} 

y_ = np.zeros(len(y))
for k in range(len(y)):
    a = y[k]
    y_[k] = int(dic_equiv[np.where(a==1)[0][0]])
print(y_)
y_ = y_.astype(int)
```

```python
# Affichage
for i in range(1, len(y), 50):
    plt.figure()
    plt.imshow(X[i].reshape(img_size, img_size))
    plt.title(y_[i])
    plt.axis('off')
```

```python
# Creation du dataframe

# Lignes correspond aux images,
# 4096 premiÃ¨res colonnes aux pixels,
# derniÃ¨re colonne Ã  la classe.
df = pd.DataFrame(X)
df['Value']=y_
print(df.head)
```
